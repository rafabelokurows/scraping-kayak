```python
import pandas as pd
import glob

#read all the csv files in the data directory
all_files = glob.glob("data/*OPO*.csv")

#create an empty dataframe
df = pd.DataFrame()
```

```python
all_files
```

```python
#loop through all the files and store them in the dataframe
for f in all_files:
    file_name = f.split("/")[-1]
    filename = file_name.split(".")[0]
    df_temp = pd.read_csv(f)
    df_temp["filename"] = filename
    df = df.append(df_temp)
```

```python
df
```

```python
#reorder the columns
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['filename'].str[:11], format='%Y%m%d%H%M')
#df['filename'].str[:11]
#datetime.strptime('20230127091', '%Y%m%d%H%M')
```

```python
Italia = df.query("Country == 'Itália'").sort_values(by=['Price'],ascending=True)
#groupby(['City']).min('Price').sort_values(by=['Price'],ascending=True)
```

```python
cols = Italia.columns.tolist()
cols = cols[-1:] + cols[:-1]
Italia = Italia[cols]
```
